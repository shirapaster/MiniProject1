```python
import numpy as np
import matplotlib.pyplot as plt
#calculating the sin function 
time=np.arange(0,1,0.01)
f=5 #frequency
y= np.sin(2*np.pi*f*time)
#creating the graph
plt.plot(time,y, label="Original Signal")

#calculating the max amd min points
max_indices = np.where(y == np.max(y))[0].tolist()
min_indices = np.where(y == np.min(y))[0].tolist()
#making the new graph indices
new_indices: list[int] = []
new_indices.extend(max_indices)
new_indices.extend(min_indices)
#re sampeling of every 5 points while maintaning the max amd min
for i in range(0,len(y),5):
    if i not in new_indices:
        new_indices.append(i)
new_indices = sorted(set(new_indices))
#creating the new signal
new_y = y[new_indices] 
new_time = time[new_indices]

#creating the new graph
plt.plot(new_time,new_y,"o", label="Downsampled Signal")
plt.title("Original vs Downsampled Sinusoidal Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.legend(['Original Signal', 'Downsampled Signal'])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('laptop_price - dataset.csv')
#Plot the price of all the laptops
plt.figure(figsize=(10, 6))
plt.scatter(df.index, df['Price (Euro)'], color='blue', alpha=0.5)
plt.title("Price of All Laptops")
plt.xlabel("Laptop Index")
plt.ylabel("Price ()")
plt.show()

#calculating the average price of each computer company
grouped_by_company=df.groupby("Company")
avg_price_by_company = grouped_by_company['Price (Euro)'].mean()
#finding the most expensive company
most_expensive_company = avg_price_by_company.idxmax()
#finding the most expensive price
highest_avg_price = avg_price_by_company.max()
#the company with the highest average price
print(f"The company with the highest average price is: {avg_price_by_company.idxmax()}")
# the highest average price of a laptop
print(f"The highest average price is: {avg_price_by_company.max()}")

#finding the unique operating systems 
unique_op_sys = df['OpSys'].unique()
print(unique_op_sys)
#checking if all the names are in the same format
df['OpSys'] = df['OpSys'].str.lower().str.strip()
# double checking!
unique_op_sys_fixed = df['OpSys'].unique()
print(unique_op_sys_fixed)
#creating the graphs!
# unique_op_sys_fixed contains the cleaned operating systems
for os in unique_op_sys_fixed: 
    os_data = df[df['OpSys'] == os]  # Filter the data for the current operating system
    plt.figure(figsize=(10, 6))  # Set the size of the plot
    plt.hist(os_data['Price (Euro)'], bins=20, alpha=0.7)  # Create a histogram for the price distribution
    plt.title(f"Price Distribution for {os.capitalize()} OS")  # Set the plot title
    plt.xlabel("Price ()")  # Set the x-axis label
    plt.ylabel("Frequency")  # Set the y-axis label
    plt.show()  # Display the plot

#What is the relationship between RAM and computer price? 
#checking that all the numbers are without errors
df['RAM (GB)'] = pd.to_numeric(df['RAM (GB)'], errors='coerce')
# creating the graph that connects between the RAM to the price 
plt.figure(figsize=(10, 6))
plt.scatter(df['RAM (GB)'], df['Price (Euro)'], alpha=0.5)
plt.title("Relationship Between RAM and Laptop Price")
plt.xlabel("RAM (GB)")
plt.ylabel("Price ()")
plt.show()
"""
Observing the scatter plot between RAM and laptop price:
From the plot, there seems to be a positive correlation between RAM size and laptop price. 
Generally, laptops with higher RAM sizes are priced higher, which aligns with the expectation 
that more RAM often indicates better performance and, therefore, a higher cost.
However, the relationship is not strictly linear. Other factors, such as brand, CPU, GPU, 
storage type, and build quality, also significantly influence the price. 
This results in some overlap in prices between laptops with similar RAM sizes.
For example, a laptop with 8GB RAM might be priced higher than one with 16GB 
if it has a more powerful processor or better display. This highlights that while RAM size 
is a strong indicator of price, it is not the sole determinant.
"""

# Extract unique values from the "Memory" column
unique_memory_values = df['Memory'].unique()
# Print all unique values to understand the data structure
print("Unique memory values:", unique_memory_values)
# Clean the 'Memory' column by converting all text to lowercase and stripping extra spaces
df['Memory'] = df['Memory'].str.lower().str.strip()

# Check the unique values in the 'Memory' column after cleaning
print(df['Memory'].unique())  # Verify that the values have been standardized

# Function to extract storage type from 'Memory'
def extract_storage_type(memory):
    # If both SSD and HDD are present, return both
    if 'ssd' in memory and 'hdd' in memory:
        return 'SSD, HDD'
    # If only SSD is present
    elif 'ssd' in memory:
        return 'SSD'
    # If only HDD is present
    elif 'hdd' in memory:
        return 'HDD'
    # If Flash Storage is present
    elif 'flash storage' in memory:
        return 'Flash Storage'
    elif 'hybrid' in memory:
        return 'hybrid'
    # For any other type not included in the conditions above
    return 'Other'

# Apply the function to the 'Memory' column and create the new 'Storage type' column
df['Storage type'] = df['Memory'].apply(extract_storage_type)

# Verify the new 'Storage type' column
print(df[['Memory', 'Storage type']].head())  # Check the first few rows of the updated DataFrame






```

```python

import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('laptop_price - dataset.csv')
"""
my first question is:
what is the most common operating system we can see in our data?
"""
# Step 1: Check the frequency of each operating system
# This will count how many times each OS appears in the 'OpSys' column
os_counts = df['OpSys'].value_counts()

# Step 2: Plot the frequency of operating systems
plt.figure(figsize=(10, 6))  # Set the size of the figure
os_counts.plot(kind='bar', color='skyblue', edgecolor='black')

# Adding labels and title
plt.title("Frequency of Operating Systems", fontsize=16)
plt.xlabel("Operating System", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability

# Show the plot
plt.tight_layout()
plt.show()
 
print("we can see that the most common operating systam is Windows 10!")

"""
My second question is: what is the impact of the operating system on the average price of laptops?
"""
# Group by 'OpSys' and calculate the average price
avg_price_by_os = df.groupby('OpSys')['Price (Euro)'].mean().sort_values(ascending=False)
# Plot the average price by operating system
plt.figure(figsize=(10, 6))  # Set figure size
avg_price_by_os.plot(kind='bar', color='lightgreen', edgecolor='black')
# Adding labels and title
plt.title("Average Laptop Price by Operating System", fontsize=16)
plt.xlabel("Operating System", fontsize=14)
plt.ylabel("Average Price ()", fontsize=14)
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability

# Show the plot
plt.tight_layout()
plt.show()

print("we can see that the highest price is related to the macOS, than Windows 10.")
```
